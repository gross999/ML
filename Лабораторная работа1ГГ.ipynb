```python
import numpy as np
import pandas as pd

dataset = pd.read_csv('carsb.csv')
dataset.head()
```

```python
exclusive = (12000, 16000)
means = dataset.groupby('CarName')['price'].mean()
filtered_means = means[means.between(exclusive[0], exclusive[1])].sort_index()
filtered_means
```
