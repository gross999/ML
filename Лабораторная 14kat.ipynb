```python
import numpy as np
import pandas as pd
from sklearn import preprocessing

dataset = pd.read_csv('new_car.csv', sep=',')
dataset.describe()
```

```python
classes = [(0, 12000), 
           (12000, 46000)]

price = dataset[['price']]
categories = []

for _, x in price.iterrows():
    doneFlag = False
    for i, c in enumerate(classes):
        if c[0] <= x['price'] < c[1]:
            categories.append(i)
            doneFlag = True
            break

    if doneFlag:
        continue

dataset['Категория'] = categories
dataset['Категория'] = dataset['Категория'].replace({1: 'yes', 0: 'no'})
dataset.head(100)
```

```python
dataset.head()
```

```python
# выберема категориальные переменные
from sklearn.compose import make_column_selector

categorical_selector = make_column_selector(dtype_include=[object])
categorical_columns = categorical_selector(dataset)
categorical_columns
```

```python
# выберем численные показатели - это все показатели
categorical_selector = make_column_selector(dtype_include=[np.float_, np.int_, int, float])
numeric_columns = categorical_selector(dataset)
numeric_columns
```

```python
df_cat = dataset[categorical_columns] #отберем только те столбцы в которых присутствует категориальная перемнная
df_cat.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder

education_column = df_cat[["Категория"]]

print(df_cat['Категория'].unique()) #посмотрим какие виды образования присутствуют в выборке

encoder = OrdinalEncoder()   # вызываем кодировщик
encoder  = encoder.fit(df_cat[['Категория']])
df_cat['Категория'] = encoder.transform(df_cat[['Категория']])
df_cat
```

```python
print(encoder.categories_)
print(len(encoder.categories_[0]))
print(df_cat['Категория'].unique())
df_cat['Категория'].unique().size
```

```python
# OneHotEncoder

from sklearn.preprocessing import OneHotEncoder

#Создаем экземпляр OneHotEncoder с параметром sparse_output
encoder = OneHotEncoder(sparse_output=False)
# Установили кодировщик в столбец "Категория" в df_cat
encoder = encoder.fit(df_cat[['Категория']])
# Преобразовали столбец "Категория", используя выученную кодировку
education_encoded = encoder.transform(df_cat[['Категория']])
education_encoded
```

```python
feature_names = encoder.get_feature_names_out(input_features=["Категория"]) 
# получаем виды образования и на их основе создаем столбцы
education_encoded = pd.DataFrame(education_encoded, columns=feature_names)
education_encoded
```

```python
df_cat_new=df_cat.join(education_encoded)
df_cat_new=df_cat_new.drop(df_cat_new[['Категория']],axis=1)
df_cat_new
```

```python
from sklearn.preprocessing import LabelEncoder
df_cat = dataset[categorical_columns] 
df_cat.head()

encoder = LabelEncoder()   # вызываем кодировщик
encoder  = encoder.fit(df_cat[['Категория']])
df_cat['Категория'] = encoder.transform(df_cat[['Категория']])
df_cat
```

```python
print(encoder.classes_)
print(len(encoder.classes_))
print(df_cat['Категория'].unique())
df_cat['Категория'].unique().size
```
